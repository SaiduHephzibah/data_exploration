```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
import matplotlib.pyplot as plt
%matplotlib tk
```

```python
dataset=pd.read_csv('data_collected.csv')
dataset.head()
```

```python
dataset.info() 
```

```python
x=dataset.iloc[:,[0,1]].values
```

```python
plt.style.use('fivethirtyeight')
```

```python
plt.subplot(211) # 2 rows 1 column position=1
plt.boxplot(x[:,0],vert=False,patch_artist=True,labels=['height']) # patch_artist =for beter visulization(fils box=blue)
plt.subplot(212)
plt.boxplot(x[:,1],vert=False,patch_artist=True,labels=['weight']) # vert=to show plot vertical or horizntal
plt.show()
```

```python
#scatter plot
plt.style.use('ggplot') 
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],s=100,color='b')
plt.xlabel('height')
plt.ylabel('weight')
plt.title('t-shirts')
plt.show()
```

```python
#identifying clusters using dendo gram
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(10,10))
dendrogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.xlabel('clusters')
plt.ylabel('distance')
plt.title('t shirts')
plt.show()
```

```python
n=5
from sklearn.cluster import AgglomerativeClustering
```

```python
hc=AgglomerativeClustering(n_clusters=n)
```

```python
y=hc.fit_predict(x)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],s=100,c=y,cmap='gnuplot')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('mall customers')
plt.show()
```
